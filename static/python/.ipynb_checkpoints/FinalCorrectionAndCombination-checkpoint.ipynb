```python
import pandas as pd
import numpy as np
```

```python
cpi = pd.read_json('../data/cpi.json', orient='index')
```

```python
cpi.describe()
```

```python
hdi = pd.read_json('../data/hdi.json', orient='index')
```

```python
hdi.describe()
```

```python
edu = pd.read_json('../data/edu.json', orient='index')
```

```python
edu.describe()
```

```python
edu = edu.reset_index()
cpi = cpi.reset_index()
```

```python
combined = cpi.merge(edu, on='index', how='left')
```

```python
hdi = hdi.reset_index()
```

```python
combined = combined.merge(hdi, on='index', how='left')
```

```python
del combined['Country']
del combined['Country_y']
```

```python
combined = combined.rename(columns={'Country_x': 'Country'})
```

```python
combined.info()
```

```python
combined[combined['HDI Score 2015'].isnull()]
# kész a tökéletes kombinált adatsor, néhány ország azért hiányzik a HDI és EDU pontokból, mert adathiány
# okául hivatalosan is kihajták őket
```

```python
combined = combined.set_index('index')
```

```python
combined.to_json('combined.json', orient='index')
```
