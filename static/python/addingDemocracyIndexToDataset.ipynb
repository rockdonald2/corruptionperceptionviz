```python
import pandas as pd
import numpy as np
```

```python
dem = pd.read_excel('dem.xlsx', sheet_name='data-for-countries-etc-by-year')
```

```python
dem = dem[dem['geo'] != 0]
```

```python
dem['geo'] = list(map(lambda x : x.upper(), dem['geo']))
```

```python
dem = dem[['geo', 'name', 'time', 'Democracy index (EIU)']]
```

```python
dem = dem.rename(columns={'geo': 'index', 'name': 'Country', 'time': 'Time', 'Democracy index (EIU)': 'Democracy Score'})
```

```python
table = dem.pivot_table('Democracy Score', ['index', 'Country'], 'Time')
```

```python
table = table.reset_index()
```

```python
table = table[['index', 'Country', 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]]
```

```python
table = table.rename(columns={2012: 'Dem Score 2012', 2013: 'Dem Score 2013', 2014: 'Dem Score 2014', 2015: 'Dem Score 2015', 2016: 'Dem Score 2016', 2017: 'Dem Score 2017', 2018: 'Dem Score 2018', 2019: 'Dem Score 2019'})
```

```python
table.set_index('index').to_json('dem.json', orient='index')
```
